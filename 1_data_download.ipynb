Librerías necesarias

```python
import pandas as pd
```

Parametros de configuración

```python
START_DATE = '2015-01-01'
PARQUET_OUTPUT_FILE = "output/assignment_parquet.parquet"
COMPLETE_PARQUET_FILE_PATH = "resources/sp500_history.parquet"
```

Leemos el parquet y mostramos la informaicón general de este

```python
complete_parquet = pd.read_parquet(COMPLETE_PARQUET_FILE_PATH)

print(f"Shape: {complete_parquet.shape}")
print(f"Rango de fechas: {complete_parquet['date'].min()} -> {complete_parquet['date'].max()}")
print(f"Tickers: {complete_parquet['symbol'].nunique()}")
complete_parquet.head(10)
```

Eliminamos columnas que tengan null (y que probablemente no nos aporten gran cosa)

```python
columns_to_drop = []
for column in complete_parquet.columns:
    null_count = complete_parquet[column].isnull().sum()
    print(f"La columna {column} tiene {null_count} nulls")
    if (null_count > 0):
        columns_to_drop.append(column)

print(f"\r\nSe borran las siguientes columnas por tener nulls: {columns_to_drop}")

complete_parquet = complete_parquet.drop(columns=columns_to_drop)
complete_parquet.head(10)
```

Eliminamos las fechas anteriores a la fecha de inicio del periodo del backtesting (datos que pesan y que no aportan)

```python
parquet = complete_parquet[complete_parquet['date'] >= START_DATE]
```

Guardamos los datos filtrados en PARQUET_OUTPUT_FILE

```python
parquet.to_parquet(PARQUET_OUTPUT_FILE, index=False)
print(f"Parquet filtrado guardado en: {PARQUET_OUTPUT_FILE}")
```
