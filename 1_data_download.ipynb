Librerías necesarias

```python
import pandas as pd
from datetime import datetime, timedelta
```

Parametros de configuración

```python
START_DATE = '2015-01-01'
PARQUET_OUTPUT_FILE = "resources/assignment_parquet.parquet"
COMPLETE_PARQUET_FILE_PATH = "resources/sp500_history.parquet"
```

Leemos el parquet y mostramos la informaicón general de este

```python
complete_parquet = pd.read_parquet(COMPLETE_PARQUET_FILE_PATH)

print(f"Shape: {complete_parquet.shape}")
print(f"Rango de fechas: {complete_parquet['date'].min()} -> {complete_parquet['date'].max()}")
print(f"Tickers: {complete_parquet['symbol'].nunique()}")
complete_parquet.head(10)
```

Eliminamos columnas que probablemente no nos aporten gran cosa

```python
columns_to_drop = ['assetid', 'security_name', 'high', 'low', 'volume', 'unadjusted_close']
print(f"\r\nSe borran las siguientes columnas: {columns_to_drop}\r\n")

complete_parquet = complete_parquet.drop(columns=columns_to_drop)
print(f"Resultado: {complete_parquet.shape}")
complete_parquet.head(10)
```

Eliminamos las fechas anteriores a la fecha de inicio del periodo del backtesting, restando 13 meses que usamos para el rebalanceo (datos que pesan y que no aportan)

```python
start = pd.Timestamp(START_DATE) - pd.DateOffset(months=13)
parquet = complete_parquet[complete_parquet['date'] >= start]
```

Guardamos los datos filtrados en PARQUET_OUTPUT_FILE

```python
parquet.to_parquet(PARQUET_OUTPUT_FILE, index=False)
print(f"Parquet filtrado guardado en: {PARQUET_OUTPUT_FILE}")
```
